```python
import pickle
# from estimations import Estimator
import pandas as pd
import numpy as np
```

```python
with open(r"data/emp_Q_shA.pkl", "rb") as input_file:
    emp_Q_shA = pickle.load(input_file)
data_adjacent = pd.read_csv('data/adjacent_zone.csv')
```

```python
def get_adjacent_zones(z):
    return data_adjacent.loc[data_adjacent['zone1'] == z]['zone2'].unique()
```

```python
## Create all states

ZONE_NUM = 263
INDEX_NUM = 95
state_list = []
for z in range(ZONE_NUM+1):
    for t in range(INDEX_NUM+1):
        state_list.append((z,t))
```

```python
## impute

imputed_emp_Q_shA = emp_Q_shA.copy()
dict_keys = imputed_emp_Q_shA.keys()
state_num = len(state_list)
for state in state_list:
    if state not in dict_keys:
        q = np.array([-10000]*state_num)
        for ad_zone in get_adjacent_zones(state[0]):
            q[ad_zone] = 0
        imputed_emp_Q_shA[state] = q
```
